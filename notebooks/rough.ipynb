```python
import numpy as np
import wandb

wandb_run = wandb.init(
        entity="DA6401",
        project="assignment-01",
        id = "Q2.1"
    )

def log_sample_images(x, y, dataset_name):

    class_names = {
        "mnist": [str(i) for i in range(10)],
        "fashion_mnist": [
            "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
            "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
        ]
    }

    table = wandb.Table(columns=[str(i) for i in range(10)])

    for cls in range(10):

        indices = np.where(y == cls)[0][:5]

        for idx in indices:
            img = x[idx]
            table.add_data(class_names[dataset_name][cls],
                           wandb.Image(img))

    wandb.log({"sample_images": table})
```

```python
from tensorflow.keras import datasets

_,(x,y) = datasets.mnist.load_data()

log_sample_images(x,y,'mnist')
```
